```python
import pandas as pd

tabela = pd.read_excel("base_dados.xlsx")
display(tabela)

faturamento = tabela ["Vendas R$ "].sum()
print(faturamento)

faturamento_setor = tabela [["Setor", "Vendas R$ "]].groupby("Setor").sum()
display(faturamento_setor)

faturamento_vendedor = tabela [["Setor", "Vendedor", "Vendas R$ "]].groupby(["Setor", "Vendedor"]).sum()
display(faturamento_vendedor)
```
